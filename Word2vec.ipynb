## Text preparation

```python
text  = text.lower()
```

```python
last_msg = re.sub('[^A-Za-z]'," ",text)
```

```python
last_msg
```

```python
sentences = sent_tokenize(last_msg)
sentence = [word_tokenize(sentence) for sentence in sentences]
```

```python
sentence
```

```python
for i in range(len(sentence)):
    sentence[i] = [word for word in sentence[i] if word not in stopwords.words('english')]
```

## Training the WORD2VEC model

```python
from gensim.models import word2vec
```

```python
model=Word2Vec(sentence,vector_size=100, window=5, min_count=1)
```
