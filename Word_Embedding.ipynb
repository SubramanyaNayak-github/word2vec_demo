```python
cv_corpus = []
for i in range(len(cv_demo)):
    review = re.sub('[^a-zA-Z0-9]',' ',cv_demo[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word)for word in review if not word in set(stopwords.words('english'))]
    review =" ".join(review)
    cv_corpus.append(review)
```
